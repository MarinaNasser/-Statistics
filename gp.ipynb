```python
import os
import shutil

# Define the source directory where the dataset is stored
source_directory = '/content/skin-burn-dataset'

# Define the destination directories for images and text files
image_destination = '/content/images'
text_destination = '/content/texts'

# Create the destination directories
os.makedirs(image_destination, exist_ok=True)
os.makedirs(text_destination, exist_ok=True)

# Iterate through the files in the source directory
'''
os.path.join(source_directory, filename): Constructs the full path to the source image file by joining the source_directory path and the filename.

os.path.join(image_destination, filename): Constructs the full path to the destination image file by joining the image_destination path and the filename.

shutil.move(source, destination): Moves the file from the source path to the destination path.
'''
for filename in os.listdir(source_directory):
    # Check if the file is an image
    if filename.endswith('.jpg') or filename.endswith('.jpeg') :
        # Move the image file to the image destination directory
        shutil.move(os.path.join(source_directory, filename), os.path.join(image_destination, filename))
    # Check if the file is a text file
    elif filename.endswith('.txt'):
        # Move the text file to the text destination directory
        shutil.move(os.path.join(source_directory, filename), os.path.join(text_destination, filename))
```

```python
# Get the list of files in the directory
image_files = os.listdir(image_destination)
text_files = os.listdir(text_destination)

# Count the number of files in the list
number_of_images = len(image_files)
number_of_text_files = len(text_files)


print(f'Number of images moved: {number_of_images}')
print(f'Number of text files moved: {number_of_text_files}')
```

```python
# Get the list of image and text file names
image_files = [filename.split('.')[0] for filename in os.listdir(image_destination) if filename.endswith('.jpg') or filename.endswith('.jpeg')]
text_files = [filename.split('.')[0] for filename in os.listdir(text_destination) if filename.endswith('.txt')]


#count number of text files without images
removed_text = 0

# Find text files without corresponding image files and remove them
for text_file in text_files:
    if text_file not in image_files:
        # Construct the file paths
        text_file_path = os.path.join(text_destination, f'{text_file}.txt')
        # Check if the file exists before attempting to remove it
        if os.path.exists(text_file_path):
            os.remove(text_file_path)
            removed_text+=1
            print(f"Removed text file: {text_file}.txt")
        else:
            print(f"Text file not found: {text_file}.txt")

print(f"Number of Text file removed: {removed_text}")
```

```python
# Count the number of files in the list
number_of_images = len(image_files)
number_of_text_files = len(text_files)

print(f'Number of images moved: {number_of_images}')
print(f'Number of text files moved: {number_of_text_files}')
```

# Splitting the dataset


```python
from sklearn.model_selection import train_test_split

# Split the data into training and validation sets
image_train, image_val, text_train, text_val = train_test_split(image_files, text_files, test_size=0.2, random_state=42)
```

```python
# Define destination directories
train_image_directory = '/content/train/images'
validation_image_directory = '/content/validation/images'
train_text_directory = '/content/train/texts'
validation_text_directory = '/content/validation/texts'

# Create destination directories
os.makedirs(train_image_directory, exist_ok=True)
os.makedirs(validation_image_directory, exist_ok=True)
os.makedirs(train_text_directory, exist_ok=True)
os.makedirs(validation_text_directory, exist_ok=True)
```

```python
# Move training images to the training directory
for image_file in image_train:
    shutil.move(os.path.join(image_destination, image_file), os.path.join('/content/train/images', image_file))

# Move validation images to the validation directory
for image_file in image_val:
    shutil.move(os.path.join(image_destination, image_file), os.path.join('/content/validation/images', image_file))

# Move training text files to the training directory
for text_file in text_train:
    shutil.move(os.path.join(text_destination, text_file), os.path.join('/content/train/texts', text_file))

# Move validation text files to the validation directory
for text_file in text_val:
    shutil.move(os.path.join(text_destination, text_file), os.path.join('/content/validation/texts', text_file))
```
